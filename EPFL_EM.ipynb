```python
with tifffile.TiffFile('Data/epfl_volumedata.tif') as tif:
    tags = tif.pages[0].tags
    meta_dict = {tag.name: tag.value for tag in tags.values()}
```

```python
epfl_meta_df = pd.DataFrame(list(meta_dict.items()), columns=['Tag', 'Value'])
epfl_meta_df.to_csv('epfl_meta_page0.csv', index=False)
```
